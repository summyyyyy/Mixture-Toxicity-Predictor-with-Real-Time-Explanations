```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import requests
from tqdm import tqdm
```

```python
file_path = "/content/drive/MyDrive/FYP/IRP/Data/Acute_Oral_Toxicity/Acute_Oral_Toxicity.xlsx"

df = pd.read_excel(file_path)

# Check data and CASRN column
df.head(3)
```

```python
print("Chemical Names count: ",len((df['Chemical Name'].unique())))
print("Unique CASRN count: ",len((df['CASRN'].unique())))
print("Total CASRN count:", len(df['CASRN']))
print("Total records:", len(df))
```

```python
# Define helper function to get SMILES from PubChem
def get_smiles_from_cas(casrn):
    """
    Given a CASRN (CAS Registry Number), return the corresponding SMILES string
    using the PubChem REST API.
    """
    if pd.isna(casrn):
        return None

    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{casrn}/property/CanonicalSMILES/TXT"
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            smiles = response.text.strip()
            # Sometimes PubChem returns multiple lines; get only first valid SMILES
            return smiles.split("\n")[0]
        else:
            return None
    except Exception as e:
        print(f"Error fetching SMILES for {casrn}: {e}")
        return None

# Apply function to CASRN column

# Add progress bar
tqdm.pandas()

# If the CASRN column name has spaces or different casing, adjust here
df['SMILES'] = df['CASRN'].progress_apply(get_smiles_from_cas)
```

```python
# Save results
output_path = "/content/drive/MyDrive/FYP/IRP/Data/Acute_Oral_Toxicity/Acute_Oral_Toxicity_with_SMILES.csv"
df.to_csv(output_path, index=False)

print(f"Done! File saved at: {output_path}")
df[['CASRN', 'SMILES']].head(10)
```
